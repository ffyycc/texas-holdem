```python
import matplotlib.pyplot as plt

# let's say this is your win-loss data
win_loss_data = [145,-150,235,340,102,170,20,-250,-300,-200,50,832,116,412,70,188,-50,-55,-700]

# cumulative win-loss
cumulative_data = [sum(win_loss_data[:i+1]) for i in range(len(win_loss_data))]

plt.figure(figsize=(10,5))
plt.plot(cumulative_data, marker='o')
plt.title('Texas Holdem Win-Loss Chart(Dollar)')
plt.xlabel('Game Number')
plt.ylabel('Cumulative Win/Loss')
plt.grid(True)
plt.show()
```
